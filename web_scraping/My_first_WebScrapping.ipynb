```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/European_Cup_and_UEFA_Champions_League_records_and_statistics'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')

print(soup)
```

```python
table = soup.find_all('table', class_ = 'wikitable sortable')[1]
```

```python
table
```

```python
headings = table.find_all('th')
```

```python
headings
```

```python
heading_titles = [head.text.strip() for head in headings]

print(heading_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = heading_titles )

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [item.text.strip() for item in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
