# The distribution of teams from each league

```python
# Group by 'League' and count the number of teams in each league
league_counts = df['Club'].groupby(df['League']).count()

# Use seaborn's color_palette with "vivid" option
colors = sns.color_palette("husl", len(league_counts))

# Plotting the histogram with custom colors
plt.figure(figsize=(10, 6))
league_counts.plot(kind='bar', color=colors)
plt.title('Number of Teams in Each League')
plt.xlabel('League')
plt.ylabel('Number of Teams')
plt.xticks(rotation=45, ha="right")
plt.show()
```

#### We can see that the most teams are from the English league (5 in total), followed by a tie between Spanish and Italian teams (3 each).

# The outcomes of matches involving teams with the most UEFA Champions League trophies

```python
selected_clubs = ["Real Madrid", "Milan", "Bayern Munich", "Liverpool"]
df_selected = df[df["Club"].isin(selected_clubs)]

# Define colors for winning, drawing, and losing
win_color = 'green'
draw_color = 'orange'
lose_color = 'red'

legend_labels = ['Win', 'Draw', 'Lose']

# Create a figure with four side-by-side pie charts
fig, axes = plt.subplots(1, len(selected_clubs), figsize=(15, 5))

# Plotting pie charts for each club
for i, club in enumerate(selected_clubs):
    ax = axes[i]
    club_data = df_selected[df_selected["Club"] == club][["W", "D", "L"]].iloc[0]
    ax.pie(club_data, labels=legend_labels, autopct='%1.1f%%', colors=[win_color, draw_color, lose_color])
    ax.set_title(club)

# Set common title for the entire figure
fig.suptitle("Teams performances with the most European title won")

plt.show()
```

#### Despite AC Milan having more UEFA Champions League trophies than Bayer or Liverpool, they win their matches in a lower percentage. Moreover, although Real Madrid has more than double the number of trophies compared to Bayern, judging by the victory percentages, it cannot be considered superior.

# Win Percentage Comparison

```python
# # Calculate win percentage for each team
win_percentage = df['W'] / df['Pld'] * 100

# Create a new DataFrame with sorted values
sorted_df = df.assign(WinPercentage=win_percentage).sort_values(by='WinPercentage', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='Club', y='WinPercentage', data=sorted_df, color='skyblue', order=sorted_df['Club'], legend=False)
plt.title('Win Percentage Comparison by Team')
plt.xlabel('Teams')
plt.ylabel('Win Percentage')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Top 10 teams with highest goal-ratio 

```python
# Calculate goal ratio for each club
df['GoalRatio'] = df['GF'] / df['Pld']

# Sort the DataFrame by goal ratio in descending order
sorted_df = df.sort_values('GoalRatio', ascending=False)

# Print only the top 10 teams with the highest goal ratio
for index in range(min(10, len(sorted_df))):
    club = sorted_df.iloc[index]['Club']
    ratio = sorted_df.iloc[index]['GoalRatio']
    print(f"{club}: {ratio:.3f}")
```

# Correlation matrix heatmap

```python
numeric_columns = df.select_dtypes(include='number')
correlation_matrix = numeric_columns.corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create a heatmap using seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)

plt.show()
```

# Teams with an Impressive Win-Loss Record: 
#### Those Winning 80% or More of Their Games

```python
# Calculate the percentage of games lost for each team
loss_percentage = (df['L'] / df['Pld']) * 100

# Check if there is a team that has lost only 10% or less
teams_with_low_loss_percentage = df.loc[loss_percentage <= 20, 'Club'].tolist()

# Print the club names if the condition is true
if teams_with_low_loss_percentage:
    print("Teams that have lost only 20% or less of their games: \n")
    for team in teams_with_low_loss_percentage:
        print(team)
else:
    print("No teams meet the condition.")
```

# Attack-Defense Balance

```python
# Calculate the goal conceded ratio for each team
goal_conceded_ratio = df['GA'] / df['Pld']

# Combine the club names, goal ratio, and goal conceded ratios
clubs_with_ratios = pd.DataFrame({
    'Club': df['Club'],
    'GoalRatio': df['GoalRatio'],
    'GoalConcededRatio': goal_conceded_ratio
})

# Sort the DataFrame based on GoalRatio in descending order and GoalConcededRatio in ascending order
sorted_clubs = clubs_with_ratios.sort_values(['GoalRatio', 'GoalConcededRatio'], ascending=[False, True])

# Select the top 5 teams
top_5_teams = sorted_clubs.head()

# Display the top 5 teams with their goal ratio and goal conceded ratio
print("Top 5 Teams with the Biggest GoalRatio and Lowest GoalConcededRatio: \n")
print(top_5_teams[['Club', 'GoalRatio', 'GoalConcededRatio']].reset_index(drop=True))

print("\n")

# Sort the DataFrame based on GoalRatio in ascending order and GoalConcededRatio in descending order
sorted_clubs2 = clubs_with_ratios.sort_values(['GoalRatio', 'GoalConcededRatio'], ascending=[True, False])

# Select the top 5 teams
top_5_teams2 = sorted_clubs2.head()

# Display the top 5 teams with their goal ratio and goal conceded ratio
print("Top 5 Teams with the Lowest GoalRatio and Highest GoalConcededRatio: \n")
print(top_5_teams2[['Club', 'GoalRatio', 'GoalConcededRatio']].reset_index(drop=True))
```

# Boxplot for distribution of collected points for different clubs

#### Considering the median, the median score is centered around 250 points. The box represents the interquartile range, i.e., the area between the lower and upper quartiles. Thus, the box contains the middle 50% of the data. The whiskers show the distribution of the data from the lower to upper quartile, where the minimum value is below 200 points and the maximum value is around 550 points. A single outlier is noticeable, representing the most successful team (Real Madrid), which is visibly distant from the points of the other teams.
